```python
import json
```

```python
# Python Dictionary
student = {
    "name": "Alice",
    "age": 20,
    "country": "USA"
}
print(type(student))
```

```python
# Convert Dictionary to JSON
json_text = json.dumps(student)

print(json_text)
```

```python
print(type(json_text))
```

```python
# Convert JSON Back to Dictionary
data = json.loads(json_text)
print(type(data))
```

```python
print(data["name"])
print(data["country"])
```

```python
# Save JSON to a File

import json
student = {
    "name": "Alice",
    "age": 20
}

with open("student.json", "w") as file:
  json.dump(student, file, indent=4)
```

```python
with open("student.json", "r") as file:
  data = json.load(file)

print(data)
```

```python
# Refactoring

import json

student = {
    "name": "Alice",
    "age": 20
}

with open("student.json", "w") as file:
    json.dump(student, file, indent=4)
```

```python
import json

def save_student(student, filename):
    with open(filename, "w") as file:
        json.dump(student, file, indent=4)

student = {
    "name": "Alice",
    "age": 20
}

save_student(student, "student.json")
```
