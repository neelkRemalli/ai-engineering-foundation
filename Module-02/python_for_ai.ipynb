```python
# Variable

name = "Alice"
age = 20
gpa = 3.8

print(name)
print(age)
print(gpa)
```

```python
# Data Types

name = "Alice"     # str
age = 20           # int
gpa = 3.8          # float
is_active = True   # bool
```

```python
# Check the type

print(type(name))
print(type(age))
print(type(gpa))
print(type(is_active))
```

```python
# Input & Output

student = input("Enter your name: ")
print("Welcome", student)
print(f"Welcome {student}")
```

```python
# Operators

a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)  # Remainder
```

```python
# Conditionals

score = 75

if score >= 50:
  print("pass")

else:
  print("Fail")
```

```python
# Loops

for i in range(5):
  print(i)
```

```python
# While Loop

count = 1

while count <= 3:
  print(count)
  count += 1
```

```python
# Lists

students = ["Alice", "Bob", "Charlie"]

print(students)
print(students[0])
```

```python
# Add a student

students.append("David")
print(students)
```

```python
# Dictionaries

students  = {
    "name": "Alice",
    "age":20,
    "gpa":3.8
}

print(students["name"])
print(students["gpa"])
```

```python
# Functions

def greet(name):
  return f"Hello, {name}"

print(greet("Alice"))
```

```python
def square(number):
  return number * number

print(square(5))
```

```python
# Basic Classes

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"My name is {self.name}.")

student = Student("Alice", 20)
student.introduce()
```

### Mini Project -- Student Manager

```python
students = []

def add_student(name, age):
  student = {
      "name": name,
      "age": age
  }

  students.append(student)

add_student("Alice", 20)
add_student("Bob", 22)


for student in students:
  print(student)
```
